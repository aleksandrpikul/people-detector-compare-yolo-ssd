```python
from pathlib import Path
p = Path("/content/src/detectors/ssd_mobilenet.py")
txt = p.read_text(encoding="utf-8")

if "def detect_people" not in txt:
    txt = txt.replace(
        "    def detect(self, frame_bgr: np.ndarray) -> List[Detection]:\n",
        "    def detect_people(self, frame_bgr: np.ndarray) -> List[Detection]:\n"
        "        return self.detect(frame_bgr)\n\n"
        "    def detect(self, frame_bgr: np.ndarray) -> List[Detection]:\n"
    )

p.write_text(txt, encoding="utf-8")
print("Patched detect_people()")
```

```python
from pathlib import Path
import re

p = Path("src/detectors/ssd_mobilenet.py")
txt = p.read_text(encoding="utf-8")

pattern = r"(self\.input_size\s*=\s*int\(input_size\)\n)"
if re.search(pattern, txt) and "self.name" not in txt:
    txt = re.sub(pattern, r"\1        self.name = \"ssd_mobilenet_v1_onnx\"\n", txt, count=1)

p.write_text(txt, encoding="utf-8")
print("Added detector.name")
```

```python
from pathlib import Path

p = Path("src/detectors/ssd_mobilenet.py")
txt = p.read_text(encoding="utf-8")

txt = txt.replace("cls=", "class_id=")

txt = txt.replace("x1=float(", "x1=int(")
txt = txt.replace("y1=float(", "y1=int(")
txt = txt.replace("x2=float(", "x2=int(")
txt = txt.replace("y2=float(", "y2=int(")

p.write_text(txt, encoding="utf-8")
print("Patched: cls -> class_id, coords -> int")
```
